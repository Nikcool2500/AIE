### Импорт библиотеки NumPy

```python
import numpy as np
```

### Примеры создания массивов:

- из обычных списков

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr1
```

- с использованием функции ones

```python
arr2 = np.ones(5)
arr2
```

### Демонстрация атрибутов:

```python
arr1.shape
```

```python
arr1.dtype
```

### Примеры индексирования и срезов:

- получение отдельного элемента

```python
arr1[0]
```

- срез списка

```python
arr1[1:5]
```

### Примеры векторных операций и базовых статистик:

- поэлементное сложение

```python
arr1 + arr2
```

- вычисление средней величины (mean) или другой простой статистики

```python
np.mean(arr1)
```
